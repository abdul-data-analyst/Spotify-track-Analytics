```python
%%sql
/* Most popular song*/
select track_name, artist, album, popularity
from spotify_tracks
order by popularity desc
limit 1;
```

```python
%%sql
/* Average of over all song*/
select AVG(popularity) as Average_popularity
from spotify_tracks
```

```python
%%sql
/* duration 5 minutes above*/
select track_name, artist, duration_minutes
from spotify_tracks
where duration_minutes > 5.0;
```

```python
%%sql
/*popular range in case statement*/
select 
    case
        when popularity >= 80 then 'Very Popular'
        when popularity >=50 then 'Popular'
        else 'Less Popular'
    end as popularity_range,
    count(*) as track_count
from spotify_tracks
group by popularity_range
```

```python
%%sql
/*longest track*/
select track_name, artist, duration_minutes
from spotify_tracks
order by duration_minutes desc
limit 1;
```

```python
%%sql
/*shortest track*/
select track_name, artist, duration_minutes
from spotify_tracks
order by duration_minutes asc
limit 1
```

```python
%%sql
/*Unique artists*/
SELECT DISTINCT artist
FROM spotify_tracks;
```

```python
%%sql
/*Average durartion of track*/
select avg(duration_minutes) as avg_duration
from spotify_tracks
```

```python
%%sql
/*Tracks with popularity > average popularity*/
select track_name, artist, popularity
from spotify_tracks
where popularity > (select avg(popularity) from spotify_tracks);
```
